```python
def gauss(x,a,x0,sigma,offset):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))+offset

def smooth(x,window_len=11,window='hanning'):
    """smooth the data using a window with requested size.
    
    This method is based on the convolution of a scaled window with the signal.
    The signal is prepared by introducing reflected copies of the signal 
    (with the window size) in both ends so that transient parts are minimized
    in the begining and end part of the output signal.
    
    input:
        x: the input signal 
        window_len: the dimension of the smoothing window; should be an odd integer
        window: the type of window from 'flat', 'hanning', 'hamming', 'bartlett', 'blackman'
            flat window will produce a moving average smoothing.

    output:
        the smoothed signal
        
    example:

    t=linspace(-2,2,0.1)
    x=sin(t)+randn(len(t))*0.1
    y=smooth(x)
    
    see also: 
    
    numpy.hanning, numpy.hamming, numpy.bartlett, numpy.blackman, numpy.convolve
    scipy.signal.lfilter
 
    TODO: the window parameter could be the window itself if an array instead of a string
    NOTE: length(output) != length(input), to correct this: return y[(window_len/2-1):-(window_len/2)] instead of just y.
    """



    if window_len<3:
        return x



    s=numpy.r_[x[window_len-1:0:-1],x,x[-2:-window_len-1:-1]]
    #print(len(s))
    if window == 'flat': #moving average
        w=numpy.ones(window_len,'d')
    else:
        w=eval('numpy.'+window+'(window_len)')

    y=numpy.convolve(w/w.sum(),s,mode='valid')
    return y

def determine_continuum(data):
    
    xs = np.linspace(0,np.shape(data)[0]+1,np.shape(data)[0])
    
    summed = np.sum(data, axis=1)
    
    smooth_signal = smooth(np.sum(data,axis=1))
    
    
    plt.clf()
    plt.plot(xs, np.sum(data, axis=1))
    plt.plot( smooth_signal)
    
    
    '''Okay now fit the OG signal with guassian centered at the smoothed values'''
    
    p0=[np.max(smooth_signal), find_nearest(smooth_signal,max(smooth_signal))[1], 10, np.median(summed)]
    
    popt, pcov = curve_fit(gauss, xs, summed, p0=p0)
    
    plt.plot(xs, gauss(xs,*popt))
    plt.show()
    
    return popt[1], 5*popt[2]

def trace(data, guess_cen):
    xs = np.linspace(0,np.shape(data)[0]+1,np.shape(data)[0])
    trace_location=[]
    plt.clf()
    plt.imshow(data)
    for j in range(np.shape(data)[1]):
        ys = data[:,j]
        smooth_signal = smooth(ys)
        idx = find_nearest(smooth_signal, max(smooth_signal))[1]
        trace_location.append(idx)
    plt.plot(trace_location)
    smooth_trace_location = smooth(trace_location)
    plt.plot(smooth_trace_location, color='yellow')
    plt.show()
    
    
    return smooth_trace_location

def fit_line(data, error, ycen, ycen1, ysize, xcen, plot, window, cr):
#fit_oiii(a[0].data, a_e[0].data, y_center, find_nearest(wave, oiii)[1])
    if cr=='yes':
        print('masking')
        data=ma.masked_where(data > 100, data)
    '''
    Create a list of fluxes from -ysize to +ysize
    '''
    
    '''
    Make sure its plotting the correct region
    '''
    print('xcen', xcen)
    try:
        y_cen = data[ycen,xcen-window:xcen+window]
    except IndexError:
        ycen = ycen1
        y_cen = data[ycen,xcen-window:xcen+window]
    #if plot=='yes':
    plt.clf()
    plt.imshow(data[ycen-ysize:ycen+ysize,xcen-window:xcen+window], vmax=50)
    plt.title('Cutout')
    plt.colorbar()
    plt.show()
    
    xs = np.linspace(0,len(y_cen)-1,len(y_cen))
    #first, fit a summed version to get a good guess on initial x pos:
    summed = np.sum(data[ycen-ysize:ycen+ysize,xcen-window:xcen+window], axis=0)
    p0=[max(summed),find_nearest(summed,max(summed))[1],5,np.median(summed)]
    popt, pcov = curve_fit(gauss, xs, summed, p0=p0)
    
    popt_total = popt
    pcov_total = pcov
    summed_total = summed
    summed_error = np.sqrt(np.sum((error[ycen-ysize:ycen+ysize,xcen-window:xcen+window])**2, axis=0))
    xs_total = xs
    
    plt.clf()
    plt.plot(xs, summed)
    plt.plot(xs, gauss(xs, *popt))
    plt.axvline(x=popt[1])

    plt.show()
    
        
    
    guess_x = popt[1]
    guess_disp = popt[2]
    
    flux_total = popt[0]*guess_disp
    flux_total_e = flux_total*np.sqrt((np.sqrt(pcov[0][0])/popt[0])**2+(np.sqrt(pcov[2][2])/popt[2])**2)
    
    print('guess x and guess disp', guess_x, guess_disp)
    if popt[2] < 0:
        
        p0=[max(summed),window,10,np.median(summed)]
        popt, pcov = curve_fit(gauss, xs, summed, p0=p0)



        plt.clf()
        plt.plot(xs, summed)
        plt.plot(xs, gauss(xs, *popt))
        plt.axvline(x=popt[1])

        plt.show()



        guess_x = popt[1]
        guess_disp = popt[2]

        flux_total = popt[0]*guess_disp
        flux_total_e = flux_total*np.sqrt((np.sqrt(pcov[0][0])/popt[0])**2+(np.sqrt(pcov[2][2])/popt[2])**2)
    
        
    print('guess x and guess disp', guess_x, guess_disp)
    popt=0
    fluxes = []
    fluxes_e = []
    height = []
    continuum = []
    print('~~~~This is the ysize~~~~~', y_size)
    for j in range(2*y_size+1):
        ys = data[ycen-y_size+j,xcen-window:xcen+window]
        
        try:
            length = len(error)
            ys_e = 1/np.sqrt(error[ycen-y_size+j,xcen-window:xcen+window])
        except TypeError:
            #if no length, then don't do the ys_e
            ys_e = 0
          
            
        try:
            p0=[max(ys),guess_x,guess_disp,np.median(ys)]
            try:
                length = len(error)
                error_key='yes'
                popt, pcov = curve_fit(gauss, xs, ys, p0=p0, sigma = ys_e)
            except TypeError:
                popt, pcov = curve_fit(gauss, xs, ys, p0=p0)
                error_key='no'
                
        except RuntimeError:
            fluxes.append(0)
            fluxes_e.append(0)
            height.append(np.median(ys))
            continuum.append(np.median(ys))
            continue
        
        if error_key=='yes' and popt[0] < np.median(ys_e):
            fluxes.append(0)
            fluxes_e.append(0)
            height.append(np.median(ys))
            continuum.append(np.median(ys))
            continue
            
        
        if plot=='yes':
            if popt[0] > 0 and popt[2] > 0 and (np.sqrt(pcov[0][0])/popt[0]) < 1 and (np.sqrt(pcov[2][2])/popt[2]) < 1:
                plt.clf()
                plt.plot(xs, ys)
                if error_key=='yes':
                    plt.plot(xs, ys_e)
                plt.plot(xs, gauss(xs, *popt))
                plt.title('Y position = '+str(j))
                plt.annotate(str(popt[0]*popt[2])+' flux ', xy=(0.1,0.06), xycoords='axes fraction')
                plt.annotate(popt, xy=(0.1,0.1), xycoords='axes fraction')
                flux_e_plot = (popt[0]*popt[2])*np.sqrt((np.sqrt(pcov[0][0])/popt[0])**2+(np.sqrt(pcov[2][2])/popt[2])**2)
                plt.annotate(str(flux_e_plot)+' error on flux', xy = (0.1, 0.02), xycoords='axes fraction')
                plt.annotate(str(np.median(ys_e))+' cont error median ', xy=(0.1, 0.2), xycoords='axes fraction')
                
                plt.show()
            
            
            
        if popt[0] < 0 or popt[2] < 0 or (np.sqrt(pcov[0][0])/popt[0]) > 1 or (np.sqrt(pcov[2][2])/popt[2]) > 1:
            fluxes.append(0)
            fluxes_e.append(0)
            height.append(popt[0])
            continuum.append(popt[3])
            continue
        flux=popt[0]*popt[2]
        flux_e=flux*np.sqrt((np.sqrt(pcov[0][0])/popt[0])**2+(np.sqrt(pcov[2][2])/popt[2])**2)
        #print('flux and flux_e for this row', flux, flux_e)
        
        fluxes.append(popt[0]*popt[2])
        fluxes_e.append(flux_e)
        height.append(popt[0])
        continuum.append(popt[3])
    return fluxes, fluxes_e, flux_total, flux_total_e, height, continuum, ycen, xs_total, summed_total, popt_total, pcov_total, summed_error
'''popt_total = popt
    pcov_total = pcov
    summed_total = summed
    xs_total = xs'''


from mpl_toolkits.axes_grid1 import make_axes_locatable
def plot_bpt(ysize, xs, ys, xs_e, ys_e, distance, name, PA, y_overall, y_overall_e, ll):
    
    
    #Kewley: log([O III]/Hβ) < 0.61/[log([N II]/Hα) − 0.05] + 1.3
    xs_kew = np.linspace(-2,0,100)
    ys_kew = [0.61/(x-0.05) + 1.3 for x in xs_kew]
    
    #Composite: 0.61/[log([N II]/Hα) − 0.47] + 1.19 > log([O III]/Hβ)
    xs_comp = np.linspace(-2,0.4,100)
    ys_comp = [0.61/(x-0.47) + 1.19 for x in xs_comp]
    
    index = distance*np.linspace(-ysize,ysize,2*ysize+1)
    plt.clf()
    fig=plt.figure(figsize=(4,4))
    ax=fig.add_subplot(111)
    
    colors = ax.scatter(xs, ys, c=index,  zorder=100, s=50)
    colors_2 = ax.scatter(xs, ll, c=index, zorder=100, s=50, marker='^')
    #ax.scatter(0, y_overall, s=50)
    #ax.axhline(y=y_overall, color='black')
    #ax.axhline(y=y_overall-y_overall_e, color='black', ls='--')
    #ax.axhline(y=y_overall+y_overall_e, color='black', ls='--')
    #ax.errorbar(0, y_overall, yerr=y_overall_e, c='black')
    cmap=plt.get_cmap('viridis')
    ecolor=cmap(1.0*index/(distance*ysize))
    ax.errorbar(xs, ys, xerr=xs_e, yerr=ys_e, c='black', fmt='o')#, c=index, cmap='Reds')
    ax.set_ylabel(r'log([OIII]$\lambda$5007/H$\beta$)', size=15)
    ax.set_xlabel(r'log([NII]$\lambda$6584/H$\alpha$)', size=15)
    ax.plot(xs_kew, ys_kew, ls='--', color='blue')
    ax.plot(xs_comp, ys_comp, color='green')
    ax.set_ylim([-1.2,1.5])
    ax.set_xlim([-2,1])
    ax.set_aspect(aspect=1)
    #plt.colorbar(colors, norm=matplotlib.colors.Normalize(vmin=-distance*ysize, vmax=distance*ysize),label='Distance from Continuum Center [$^{\prime\prime}$]')
    
    
    divider = make_axes_locatable(ax)
    cax = divider.new_horizontal(size="5%", pad=0.1)#, pad=0.7, pack_start=True)
    fig.add_axes(cax)
    #fig.colorbar(colors, cax=cax,orientation="vertical")
    cb1 = matplotlib.colorbar.ColorbarBase(cax, cmap='viridis',
                                norm=matplotlib.colors.Normalize(vmin=-distance*ysize, vmax=distance*ysize),
                                orientation='vertical', label = 'Distance from Continuum Center [$^{\prime\prime}$]')
    '''axbar=fig.add_subplot(122)
    cb1 = matplotlib.colorbar.ColorbarBase(axbar, cmap='viridis',
                                norm=matplotlib.colors.Normalize(vmin=-distance*ysize, vmax=distance*ysize),
                                orientation='vertical')'''
    #cb1.set_label('Distance from Continuum Center [$^{\prime\prime}$]')
    plt.tight_layout()
    plt.savefig('Adi_BPT/'+str(name)+'_PA_'+str(PA)+'.png', dpi=500)   
    
  
def summed_plot_SDSS( name, SDSS):
    
    
    
    plt.clf()
    fig=plt.figure(figsize=(4,4))
    ax=fig.add_subplot(111)
    
    if SDSS==0:
        print('no SDSS')
    else:
        '''error_hbeta = popt_hbeta[0]*popt_hbeta[2]*np.sqrt((np.sqrt(pcov_hbeta[0][0])/popt_hbeta[0])**2+(np.sqrt(pcov_hbeta[2][2])/popt_hbeta[2])**2)
    hbeta_flux = popt_hbeta[0]*popt_hbeta[2]
    ax1.annotate('Flux = '+str(round(hbeta_flux,1))+' +/- '+str(round(error_hbeta,1)), xy=(0.02,0.87), xycoords='axes fraction')
    
    ax1.set_title(r'H$\beta$')
    
    
    ratio_y = np.log10(np.array(popt_oiii[0]*popt_oiii[2])/np.array(popt_hbeta[0]*popt_hbeta[2]))
    #ratio_y_ll = np.log10(np.array(oiii_fit[4])/np.array(hbeta_fit[5]))
    dy = np.sqrt((np.array(error_oiii)/np.array(oiii_flux))**2+(np.array(error_hbeta)/np.array(hbeta_flux))**2)
    ratio_y_e = 0.434*(dy)'''
        hbeta_flux_SDSS = SDSS[0][0]*SDSS[0][2]
        error_hbeta_SDSS = SDSS[0][0]*SDSS[0][2]*np.sqrt((np.sqrt(SDSS[0][1])/SDSS[0][0])**2+(np.sqrt(SDSS[0][3])/SDSS[0][2])**2)
        
        oiii_flux_SDSS = SDSS[1][0]*SDSS[1][2]
        error_oiii_SDSS = SDSS[1][0]*SDSS[1][2]*np.sqrt((np.sqrt(SDSS[1][1])/SDSS[1][0])**2+(np.sqrt(SDSS[1][3])/SDSS[1][2])**2)
        ratio_SDSS = np.log10(np.array(oiii_flux_SDSS/hbeta_flux_SDSS))
        dy_SDSS = np.sqrt((np.array(error_oiii_SDSS)/np.array(oiii_flux_SDSS))**2+(np.array(error_hbeta_SDSS)/np.array(hbeta_flux_SDSS))**2)
        ratio_y_e_SDSS = 0.434*(dy_SDSS)
        
        halpha_flux_SDSS = SDSS[2][0]*SDSS[2][2]
        error_halpha_SDSS = SDSS[2][0]*SDSS[2][2]*np.sqrt((np.sqrt(SDSS[2][1])/SDSS[2][0])**2+(np.sqrt(SDSS[2][3])/SDSS[2][2])**2)
        
        nii_flux_SDSS = SDSS[3][0]*SDSS[3][2]
        error_nii_SDSS = SDSS[3][0]*SDSS[3][2]*np.sqrt((np.sqrt(SDSS[3][1])/SDSS[3][0])**2+(np.sqrt(SDSS[3][3])/SDSS[3][2])**2)
        ratio_SDSS_nii_halpha = np.log10(np.array(nii_flux_SDSS/halpha_flux_SDSS))
        dx_SDSS = np.sqrt((np.array(error_nii_SDSS)/np.array(nii_flux_SDSS))**2+(np.array(error_halpha_SDSS)/np.array(halpha_flux_SDSS))**2)
        ratio_x_e_SDSS = 0.434*(dx_SDSS)
        
        ax.scatter(ratio_SDSS_nii_halpha, ratio_SDSS,  zorder=100, s=5, c='orange', label='SDSS')
        ax.errorbar(ratio_SDSS_nii_halpha, ratio_SDSS, yerr=ratio_y_e_SDSS, xerr = ratio_x_e_SDSS, c='orange',fmt='none')
        print('ratio sdss', ratio_SDSS_nii_halpha, ratio_x_e_SDSS, ratio_SDSS, ratio_y_e_SDSS)
 
    
    #Kewley: log([O III]/Hβ) < 0.61/[log([N II]/Hα) − 0.05] + 1.3
    xs_kew = np.linspace(-2,0,100)
    ys_kew = [0.61/(x-0.05) + 1.3 for x in xs_kew]
    
    #Composite: 0.61/[log([N II]/Hα) − 0.47] + 1.19 > log([O III]/Hβ)
    xs_comp = np.linspace(-2,0.4,100)
    ys_comp = [0.61/(x-0.47) + 1.19 for x in xs_comp]
    ax.plot(xs_kew, ys_kew, ls='--', color='black')
    ax.plot(xs_comp, ys_comp, color='black')
    ax.set_ylim([-1.2,1.5])
    ax.set_xlim([-2,1])
    ax.set_aspect(aspect=1)
    
    ax.set_ylabel(r'log([OIII]$\lambda$5007/H$\beta$)', size=10)
    ax.set_xlabel(r'log([NII]$\lambda$6584/H$\alpha$)', size=10)
    
    #ax.axhline(y=1, ls='--', color='black')
    #ax.axhline(y=1, ls='--', color='black')
    
    
    ##ax.set_ylim([-1.2,1.5])
    ##ax.set_xlim([-distance*ysize,distance*ysize])
    
    plt.legend()
    
    plt.tight_layout()
    plt.title(str(name))
    plt.savefig('Adi_BPT/'+str(name)+'_summed_BPT.png', dpi=500) 
    



def summed_plot( name, PA, y_overall, y_overall_e, popt_oiii, popt_hbeta, SDSS, spec, ratio_blue, ratio_blue_e, ratio_red, ratio_red_e):
    
    
    
    plt.clf()
    fig=plt.figure(figsize=(4,4))
    ax=fig.add_subplot(111)
    
    if SDSS==0:
        print('no SDSS')
    else:
        '''error_hbeta = popt_hbeta[0]*popt_hbeta[2]*np.sqrt((np.sqrt(pcov_hbeta[0][0])/popt_hbeta[0])**2+(np.sqrt(pcov_hbeta[2][2])/popt_hbeta[2])**2)
    hbeta_flux = popt_hbeta[0]*popt_hbeta[2]
    ax1.annotate('Flux = '+str(round(hbeta_flux,1))+' +/- '+str(round(error_hbeta,1)), xy=(0.02,0.87), xycoords='axes fraction')
    
    ax1.set_title(r'H$\beta$')
    
    
    ratio_y = np.log10(np.array(popt_oiii[0]*popt_oiii[2])/np.array(popt_hbeta[0]*popt_hbeta[2]))
    #ratio_y_ll = np.log10(np.array(oiii_fit[4])/np.array(hbeta_fit[5]))
    dy = np.sqrt((np.array(error_oiii)/np.array(oiii_flux))**2+(np.array(error_hbeta)/np.array(hbeta_flux))**2)
    ratio_y_e = 0.434*(dy)'''
        hbeta_flux_SDSS = SDSS[0][0]*SDSS[0][2]
        error_hbeta_SDSS = SDSS[0][0]*SDSS[0][2]*np.sqrt((np.sqrt(SDSS[0][1])/SDSS[0][0])**2+(np.sqrt(SDSS[0][3])/SDSS[0][2])**2)
        
        oiii_flux_SDSS = SDSS[1][0]*SDSS[1][2]
        error_oiii_SDSS = SDSS[1][0]*SDSS[1][2]*np.sqrt((np.sqrt(SDSS[1][1])/SDSS[1][0])**2+(np.sqrt(SDSS[1][3])/SDSS[1][2])**2)
        ratio_SDSS = np.log10(np.array(oiii_flux_SDSS/hbeta_flux_SDSS))
        dy_SDSS = np.sqrt((np.array(error_oiii_SDSS)/np.array(oiii_flux_SDSS))**2+(np.array(error_hbeta_SDSS)/np.array(hbeta_flux_SDSS))**2)
        ratio_y_e_SDSS = 0.434*(dy_SDSS)
        
        halpha_flux_SDSS = SDSS[2][0]*SDSS[2][2]
        error_halpha_SDSS = SDSS[2][0]*SDSS[2][2]*np.sqrt((np.sqrt(SDSS[2][1])/SDSS[2][0])**2+(np.sqrt(SDSS[2][3])/SDSS[2][2])**2)
        
        nii_flux_SDSS = SDSS[3][0]*SDSS[3][2]
        error_nii_SDSS = SDSS[3][0]*SDSS[3][2]*np.sqrt((np.sqrt(SDSS[3][1])/SDSS[3][0])**2+(np.sqrt(SDSS[3][3])/SDSS[3][2])**2)
        ratio_SDSS_nii_halpha = np.log10(np.array(nii_flux_SDSS/halpha_flux_SDSS))
        dx_SDSS = np.sqrt((np.array(error_nii_SDSS)/np.array(nii_flux_SDSS))**2+(np.array(error_halpha_SDSS)/np.array(halpha_flux_SDSS))**2)
        ratio_x_e_SDSS = 0.434*(dx_SDSS)
        
        ax.scatter(ratio_SDSS_nii_halpha, ratio_SDSS,  zorder=100, s=5, c='orange', label='SDSS')
        ax.errorbar(ratio_SDSS_nii_halpha, ratio_SDSS, yerr=ratio_y_e_SDSS, xerr = ratio_x_e_SDSS, c='orange',fmt='none')
        print('SDSS coords', 'nii_halpha', ratio_SDSS_nii_halpha, ratio_x_e_SDSS, 'oiii_hbeta', ratio_SDSS, ratio_y_e_SDSS)
    if y_overall_e > 1:
        print(popt_oiii, popt_hbeta)
        ll = (popt_oiii[0]-popt_oiii[3])/popt_hbeta[3]
        ratio_ll = np.log10(np.array(ll))
        print(np.shape(ratio_ll))
        ax.scatter(0, ratio_ll, s=5, marker='^', label='Lower Limit PA '+str(PA), zorder=100, color='black')
        ax.errorbar(0, ratio_ll,  c='black',fmt='none')#, c=index, cmap='Reds')
        print('lower limit', ratio_ll)
    
    else:
        ax.scatter(0, y_overall,  zorder=100, s=5, label='PA '+str(PA), color='black')
    
    
    
    
        ax.errorbar(0, y_overall, yerr=y_overall_e,  c='black',fmt='none')#, c=index, cmap='Reds')
        print('overall ratio', y_overall, y_overall_e)
    #ax.scatter(0, y_overall, c='black')
    #ax.errorbar(0, y_overall, yerr=y_overall_e, c='black')
    #ax.axhline(y=y_overall, color='black')
    #ax.axhline(y=y_overall-y_overall_e, color='black', ls='--')
    #ax.axhline(y=y_overall+y_overall_e, color='black', ls='--')
    
    #ax.xaxis.set_ticklabels([])
    if spec=='yes':
        #then additionally plot the left and right side of the fit
        ax.scatter(0, ratio_blue,  zorder=100, s=5, label='Blue PA '+str(PA), color='blue')
    
    
    
    
        ax.errorbar(0, ratio_blue, yerr=ratio_blue_e, c='blue',fmt='none')#, c=index, cmap='Reds')
        
        ax.scatter(0, ratio_red,  zorder=100, s=5, label='Red PA '+str(PA), color='red')
    
    
    
    
        ax.errorbar(0, ratio_red, yerr=ratio_red_e,  c='red',fmt='none')#, c=index, cmap='Reds')
        
        print('blue', ratio_blue, ratio_blue_e)
        print('red', ratio_red, ratio_red_e)
    #Kewley: log([O III]/Hβ) < 0.61/[log([N II]/Hα) − 0.05] + 1.3
    xs_kew = np.linspace(-2,0,100)
    ys_kew = [0.61/(x-0.05) + 1.3 for x in xs_kew]
    
    #Composite: 0.61/[log([N II]/Hα) − 0.47] + 1.19 > log([O III]/Hβ)
    xs_comp = np.linspace(-2,0.4,100)
    ys_comp = [0.61/(x-0.47) + 1.19 for x in xs_comp]
    ax.plot(xs_kew, ys_kew, ls='--', color='black')
    ax.plot(xs_comp, ys_comp, color='black')
    ax.set_ylim([-1.2,1.5])
    ax.set_xlim([-2,1])
    ax.set_aspect(aspect=1)
    
    ax.set_ylabel(r'log([OIII]$\lambda$5007/H$\beta$)', size=10)
    ax.set_xlabel(r'log([NII]$\lambda$6584/H$\alpha$)', size=10)
    
    ax.axhline(y=1, ls='--', color='black')
    #ax.axhline(y=1, ls='--', color='black')
    
    
    ##ax.set_ylim([-1.2,1.5])
    ##ax.set_xlim([-distance*ysize,distance*ysize])
    
    plt.legend()
    
    plt.tight_layout()
    plt.title(str(name))
    plt.savefig('Adi_BPT/'+str(name)+'_PA_'+str(PA)+'_summed_BPT.png', dpi=500) 
    
    return
    
    #same thing but no x axis or x error bars
    
    plt.clf()
    fig=plt.figure(figsize=(4,4))
    ax=fig.add_subplot(111)
    
    if SDSS==0:
        print('no SDSS')
    else:
        '''error_hbeta = popt_hbeta[0]*popt_hbeta[2]*np.sqrt((np.sqrt(pcov_hbeta[0][0])/popt_hbeta[0])**2+(np.sqrt(pcov_hbeta[2][2])/popt_hbeta[2])**2)
    hbeta_flux = popt_hbeta[0]*popt_hbeta[2]
    ax1.annotate('Flux = '+str(round(hbeta_flux,1))+' +/- '+str(round(error_hbeta,1)), xy=(0.02,0.87), xycoords='axes fraction')
    
    ax1.set_title(r'H$\beta$')
    
    
    ratio_y = np.log10(np.array(popt_oiii[0]*popt_oiii[2])/np.array(popt_hbeta[0]*popt_hbeta[2]))
    #ratio_y_ll = np.log10(np.array(oiii_fit[4])/np.array(hbeta_fit[5]))
    dy = np.sqrt((np.array(error_oiii)/np.array(oiii_flux))**2+(np.array(error_hbeta)/np.array(hbeta_flux))**2)
    ratio_y_e = 0.434*(dy)'''
        hbeta_flux_SDSS = SDSS[0][0]*SDSS[0][2]
        error_hbeta_SDSS = SDSS[0][0]*SDSS[0][2]*np.sqrt((np.sqrt(SDSS[0][1])/SDSS[0][0])**2+(np.sqrt(SDSS[0][3])/SDSS[0][2])**2)
        
        oiii_flux_SDSS = SDSS[1][0]*SDSS[1][2]
        error_oiii_SDSS = SDSS[1][0]*SDSS[1][2]*np.sqrt((np.sqrt(SDSS[1][1])/SDSS[1][0])**2+(np.sqrt(SDSS[1][3])/SDSS[1][2])**2)
        ratio_SDSS = np.log10(np.array(oiii_flux_SDSS/hbeta_flux_SDSS))
        dy_SDSS = np.sqrt((np.array(error_oiii_SDSS)/np.array(oiii_flux_SDSS))**2+(np.array(error_hbeta_SDSS)/np.array(hbeta_flux_SDSS))**2)
        ratio_y_e_SDSS = 0.434*(dy_SDSS)
        
        halpha_flux_SDSS = SDSS[2][0]*SDSS[2][2]
        error_halpha_SDSS = SDSS[2][0]*SDSS[2][2]*np.sqrt((np.sqrt(SDSS[2][1])/SDSS[2][0])**2+(np.sqrt(SDSS[2][3])/SDSS[2][2])**2)
        
        nii_flux_SDSS = SDSS[3][0]*SDSS[3][2]
        error_nii_SDSS = SDSS[3][0]*SDSS[3][2]*np.sqrt((np.sqrt(SDSS[3][1])/SDSS[3][0])**2+(np.sqrt(SDSS[3][3])/SDSS[3][2])**2)
        ratio_SDSS_nii_halpha = np.log10(np.array(nii_flux_SDSS/halpha_flux_SDSS))
        dx_SDSS = np.sqrt((np.array(error_nii_SDSS)/np.array(nii_flux_SDSS))**2+(np.array(error_halpha_SDSS)/np.array(halpha_flux_SDSS))**2)
        ratio_x_e_SDSS = 0.434*(dx_SDSS)
        
        ax.scatter(-1, ratio_SDSS,  zorder=100, s=5, c='orange', label='SDSS')
        ax.errorbar(-1, ratio_SDSS, yerr=ratio_y_e_SDSS,  c='orange',fmt='none')
        
    if y_overall_e > 1:
        print(popt_oiii, popt_hbeta)
        ll = (popt_oiii[0]-popt_oiii[3])/popt_hbeta[3]
        ratio_ll = np.log10(np.array(ll))
        print(np.shape(ratio_ll))
        ax.scatter(0, ratio_ll, s=5, marker='^', label='Lower Limit PA '+str(PA), color='black')
        ax.errorbar(0, ratio_ll, c='black',fmt='none')#, c=index, cmap='Reds')
    
    else:
        ax.scatter(0, y_overall,  zorder=100, s=5, label='PA '+str(PA), color='black')
    
    
    
    
        ax.errorbar(0, y_overall, yerr=y_overall_e,  c='black',fmt='none')#, c=index, cmap='Reds')
    
    if spec=='yes':
        #then additionally plot the left and right side of the fit
        ax.scatter(1, ratio_blue,  zorder=100, s=5, label='Blue PA '+str(PA), color='blue')
    
    
    
    
        ax.errorbar(1, ratio_blue, yerr=ratio_blue_e, c='blue',fmt='none')#, c=index, cmap='Reds')
        
        ax.scatter(1, ratio_red,  zorder=100, s=5, label='Red PA '+str(PA), color='red')
    
    
    
    
        ax.errorbar(1, ratio_red, yerr=ratio_blue_e, c='red',fmt='none')#, c=index, cmap='Reds')
    
    
    ax.xaxis.set_ticklabels([])
    
    ax.set_ylabel(r'log([OIII]$\lambda$5007/H$\beta$)', size=10)
    #ax.set_xlabel(r'Blank Axis', size=10)
    ax.set_ylim([-1.2,1.5])
    ax.axhline(y=1, ls='--', color='black')
    #ax.axhline(y=1, ls='--', color='black')
    
    
    ##ax.set_ylim([-1.2,1.5])
    ##ax.set_xlim([-distance*ysize,distance*ysize])
    
    plt.legend()
    
    plt.tight_layout()
    plt.title(str(name))
    ax.set_xlim([-3,3])
    plt.savefig('Adi_BPT/'+str(name)+'_PA_'+str(PA)+'_summed.png', dpi=500)
    
def spatial_plot(ysize, ys, ys_e, distance, name, PA, y_overall, y_overall_e, ll):
    
    
    
    index = distance*np.linspace(-ysize,ysize,2*ysize+1)
    
    
    plt.clf()
    fig=plt.figure(figsize=(4,4))
    ax=fig.add_subplot(111)
    
    #color all AGN points purple
    color_points=[]
    for j in range(len(ys)):
        if (ys[j] - ys_e[j])>1:
            color_points.append('purple')
        else:
            color_points.append('black')
    color_points_ll=[]
    for j in range(len(ll)):
        if ll[j]>1:
            color_points_ll.append('purple')
        else:
            color_points_ll.append('black')        
    
    colors = ax.scatter(index, ys, c=color_points,  zorder=100, s=50)
    
    print('~~~index, ys, ys_e~~~')
    for j in range(len(index)):
        print(index[j], ys[j], ys_e[j])
    
    colors = ax.scatter(index, ll, c=color_points_ll, zorder=100, s=50, marker='^')
    cmap=plt.get_cmap('viridis')
    
    ax.errorbar(index, ys, yerr=ys_e, c='black',fmt='o')#, c=index, cmap='Reds')
    
    #ax.scatter(0, y_overall, c='black')
    #ax.errorbar(0, y_overall, yerr=y_overall_e, c='black')
    #ax.axhline(y=y_overall, color='black')
    #ax.axhline(y=y_overall-y_overall_e, color='black', ls='--')
    #ax.axhline(y=y_overall+y_overall_e, color='black', ls='--')
    
    ax.set_ylabel(r'log([OIII]$\lambda$5007/H$\beta$)', size=10)
    ax.set_xlabel(r'Distance from Continuum Center [$^{\prime\prime}$]', size=10)
    ax.axvline(x=0, ls='--', color='black')
    #ax.axhline(y=1, ls='--', color='black')
    
    
    ##ax.set_ylim([-1.2,1.5])
    ##ax.set_xlim([-distance*ysize,distance*ysize])
    
    
    
    #ax.set_aspect(aspect=1)
    #plt.colorbar(colors, norm=matplotlib.colors.Normalize(vmin=-distance*ysize, vmax=distance*ysize),label='Distance from Continuum Center [$^{\prime\prime}$]')
    
    
    '''divider = make_axes_locatable(ax)
    cax = divider.new_horizontal(size="5%", pad=0.1)#, pad=0.7, pack_start=True)
    fig.add_axes(cax)
    #fig.colorbar(colors, cax=cax,orientation="vertical")
    cb1 = matplotlib.colorbar.ColorbarBase(cax, cmap='viridis',
                                norm=matplotlib.colors.Normalize(vmin=-distance*ysize, vmax=distance*ysize),
                                orientation='vertical', label = 'Distance from Continuum Center [$^{\prime\prime}$]')
    '''
    #cb1.set_label('Distance from Continuum Center [$^{\prime\prime}$]')
    plt.tight_layout()
    plt.savefig('Adi_BPT/'+str(name)+'_PA_'+str(PA)+'_spatial.png', dpi=500)   
    
    
def two_d_plot(data, trace_cen_oiii, trace_cen_hbeta, og_y_cen, y_size, xcen_oiii, xcen_hbeta, PA, name, pixelscale, oiii_lambda, hbeta_lambda, CDELT):
    #two_d_plot(a[0].data, trace_line[find_nearest(wave, oiii)[1]],y_center,y_size, find_nearest(wave, oiii)[1], find_nearest(wave, hbeta)[1])
    print(trace_cen_oiii, trace_cen_hbeta)
    
    data_cutout=data[int(trace_cen_oiii)-y_size:int(trace_cen_oiii)+y_size,xcen_oiii-50:xcen_oiii+50]
    
    plt.clf()
    fig=plt.figure()
    ax0 = fig.add_subplot(211)
    im0 = ax0.imshow(data_cutout)#, vmax=data_cutout.max()-50, vmin=np.median(data_cutout))
    plt.colorbar(im0)
    ax0.set_title(r'[OIII]$\lambda$5007')
    ax0.yaxis.set_ticks([0,y_size,2*y_size])
    ax0.yaxis.set_ticklabels([-round(y_size*pixelscale,2), 0, round(y_size*pixelscale,2)])
    
    
    xticks = [0, 10, 20, 30, 40, 50, 60]
    xlabels = [round((x-30)*CDELT+oiii_lambda,2) for x in xticks]
    ax0.xaxis.set_ticks(xticks)
    ax0.xaxis.set_ticklabels(xlabels)
    
    data_cutout=data[int(trace_cen_hbeta)-y_size:int(trace_cen_hbeta)+y_size,xcen_hbeta-50:xcen_hbeta+50]
    ax0.axhline(y=y_size, lw=0.2, color='white')
    
    ax1 = fig.add_subplot(212)
    im1 = ax1.imshow(data_cutout)#, vmax=max(data_cutout)-50, vmin=np.median(data_cutout))
    plt.colorbar(im1)
    ax1.set_title(r'H$\beta$')
    ax1.yaxis.set_ticks([0,y_size,2*y_size])
    ax1.yaxis.set_ticklabels([-round(y_size*pixelscale,2), 0, round(y_size*pixelscale,2)])
    ax1.axhline(y=y_size, lw=0.2, color='white')
    plt.savefig('Adi_BPT/'+str(name)+'_PA_'+str(PA)+'_2D.png', dpi=500)
    
def plot_line_fit(name, PA, xs_oiii, sum_oiii, sum_oiii_e, popt_oiii, pcov_oiii, xs_hbeta, sum_hbeta, sum_hbeta_e, popt_hbeta, pcov_hbeta, spec):
#plot_line_fit(gal, PA1, oiii_fit[8], oiii_fit[9], oiii_fit[10], oiii_fit[11], hbeta_fit[8], hbeta_fit[9], hbeta_fit[10], hbeta_fit[11])
    
    if spec =='yes':
        #then we need to fit two gaussians and return it
        p2 = [popt_oiii[0], popt_oiii[1]-5, popt_oiii[2]/3, popt_oiii[3], popt_oiii[0], popt_oiii[1]+5, popt_oiii[2]/3]
        print('this was the 1 gaussian fit', popt_oiii)
        print('this is our guess', p2)
        popt2, pcov2 = curve_fit(gauss2, xs_oiii, sum_oiii, p0=p2, sigma = sum_oiii_e)
        print('popt2 oiii', popt2)
        error_oiii_1 = popt2[0]*popt2[2]*np.sqrt((np.sqrt(pcov2[0][0])/popt2[0])**2+(np.sqrt(pcov2[2][2])/popt2[2])**2)
        oiii_flux_1 = popt2[0]*popt2[2]
        
        error_oiii_2 = popt2[4]*popt2[6]*np.sqrt((np.sqrt(pcov2[4][4])/popt2[4])**2+(np.sqrt(pcov2[6][6])/popt2[6])**2)
        oiii_flux_2 = popt2[4]*popt2[6]
        
        
        
        
        plt.clf()
        fig = plt.figure()
        ax0 = fig.add_subplot(211)
        ax0.scatter(xs_oiii, sum_oiii, color='black')
        ax0.plot(xs_oiii, gauss(xs_oiii, *popt_oiii), color='black')
        
        gauss_1_p0 = [popt2[0], popt2[1], popt2[2], popt2[3]]
        gauss_2_p0 = [popt2[4], popt2[5], popt2[6], popt2[3]]
        ax0.plot(xs_oiii, gauss(xs_oiii, *gauss_1_p0), color='blue')
        ax0.plot(xs_oiii, gauss(xs_oiii, *gauss_2_p0), color='red')
        ax0.set_title(r'[OIII]$\lambda$5007')
        error_oiii = popt_oiii[0]*popt_oiii[2]*np.sqrt((np.sqrt(pcov_oiii[0][0])/popt_oiii[0])**2+(np.sqrt(pcov_oiii[2][2])/popt_oiii[2])**2)
        oiii_flux = popt_oiii[0]*popt_oiii[2]
        #ax0.annotate('Flux = '+str(round(oiii_flux,1))+' +/- '+str(round(error_oiii,1)), xy=(0.02,0.87), xycoords='axes fraction')


        
        p2 = [popt_hbeta[0], popt_hbeta[1]-5, popt_hbeta[2]/3, popt_hbeta[3], popt_hbeta[0], popt_hbeta[1]+5, popt_hbeta[2]/3]
        popt2, pcov2 = curve_fit(gauss2, xs_hbeta, sum_hbeta, p0=p2, sigma = sum_hbeta_e)
        print('popt2 hbeta', popt2)
        
        error_hbeta_1 = abs(popt2[0]*popt2[2])*np.sqrt((np.sqrt(pcov2[0][0])/popt2[0])**2+(np.sqrt(pcov2[2][2])/popt2[2])**2)
        hbeta_flux_1 = abs(popt2[0]*popt2[2])
        
        error_hbeta_2 = abs(popt2[4]*popt2[6])*np.sqrt((np.sqrt(pcov2[4][4])/popt2[4])**2+(np.sqrt(pcov2[6][6])/popt2[6])**2)
        hbeta_flux_2 = abs(popt2[4]*popt2[6])
        
        ratio_y_1 = np.log10(np.array(oiii_flux_1)/np.array(hbeta_flux_1))
        #ratio_y_ll = np.log10(np.array(oiii_fit[4])/np.array(hbeta_fit[5]))
        dy_1 = np.sqrt((np.array(error_oiii_1)/np.array(oiii_flux_1))**2+(np.array(error_hbeta_1)/np.array(hbeta_flux_1))**2)
        ratio_y_e_1 = 0.434*(dy_1)
        
        ratio_y_2 = np.log10(np.array(oiii_flux_2)/np.array(hbeta_flux_2))
        #ratio_y_ll = np.log10(np.array(oiii_fit[4])/np.array(hbeta_fit[5]))
        dy_2 = np.sqrt((np.array(error_oiii_2)/np.array(oiii_flux_2))**2+(np.array(error_hbeta_2)/np.array(hbeta_flux_2))**2)
        ratio_y_e_2 = 0.434*(dy_2)
        

        ax1 = fig.add_subplot(212)
        ax1.scatter(xs_hbeta, sum_hbeta, color='black')
        ax1.plot(xs_hbeta, gauss(xs_hbeta, *popt_hbeta), color='black')
        gauss_1_p0 = [popt2[0], popt2[1], popt2[2], popt2[3]]
        gauss_2_p0 = [popt2[4], popt2[5], popt2[6], popt2[3]]
        ax1.plot(xs_hbeta, gauss(xs_hbeta, *gauss_1_p0), color='blue')
        ax1.plot(xs_hbeta, gauss(xs_hbeta, *gauss_2_p0), color='red')
        
        
        error_hbeta = popt_hbeta[0]*popt_hbeta[2]*np.sqrt((np.sqrt(pcov_hbeta[0][0])/popt_hbeta[0])**2+(np.sqrt(pcov_hbeta[2][2])/popt_hbeta[2])**2)
        hbeta_flux = popt_hbeta[0]*popt_hbeta[2]
        #ax1.annotate('Flux = '+str(round(hbeta_flux,1))+' +/- '+str(round(error_hbeta,1)), xy=(0.02,0.87), xycoords='axes fraction')

        ax1.set_title(r'H$\beta$')


        ratio_y = np.log10(np.array(popt_oiii[0]*popt_oiii[2])/np.array(popt_hbeta[0]*popt_hbeta[2]))
        #ratio_y_ll = np.log10(np.array(oiii_fit[4])/np.array(hbeta_fit[5]))
        dy = np.sqrt((np.array(error_oiii)/np.array(oiii_flux))**2+(np.array(error_hbeta)/np.array(hbeta_flux))**2)
        ratio_y_e = 0.434*(dy)


        ax0.annotate(r'log(OIII/H$\beta$) = '+str(round(ratio_y,2))+' +/- '+str(round(ratio_y_e,2)), xy=(0.55, 0.87), xycoords='axes fraction')
        ax0.annotate(r'Blue log(OIII/H$\beta$) = '+str(round(ratio_y_1,2))+' +/- '+str(round(ratio_y_e_1,2)), xy=(0.55, 0.77), xycoords='axes fraction', color='blue')
        ax0.annotate(r'Red log(OIII/H$\beta$) = '+str(round(ratio_y_2,2))+' +/- '+str(round(ratio_y_e_2,2)), xy=(0.55, 0.67), xycoords='axes fraction', color='red')

        if error_oiii > oiii_flux or error_hbeta > hbeta_flux:
            #then plot the lower limit
            ax1.axhline(y=popt_hbeta[3], color='red')
            ll = (popt_oiii[0]-popt_oiii[3])/popt_hbeta[3]
            ratio_ll = np.log10(np.array(ll))
            ax0.annotate('Lower limit = '+str(round(ratio_ll,2)), xy=(0.55, 0.77), xycoords='axes fraction')

        plt.tight_layout()
        ax0.set_ylim([np.min(sum_oiii)-0.2*np.max(sum_oiii),np.max(sum_oiii)+0.2*np.max(sum_oiii)])
        ax1.set_ylim([np.min(sum_hbeta)-0.2*np.max(sum_hbeta),np.max(sum_hbeta)+0.2*np.max(sum_hbeta)])

        plt.savefig('Adi_BPT/'+str(name)+'_PA_'+str(PA)+'_gaussian.png', dpi=500)
        return ratio_y_1, ratio_y_e_1, ratio_y_2, ratio_y_e_2
    else:
        plt.clf()
        fig = plt.figure()
        ax0 = fig.add_subplot(211)
        ax0.scatter(xs_oiii, sum_oiii, color='black')
        ax0.plot(xs_oiii, gauss(xs_oiii, *popt_oiii), color='black')
        ax0.set_title(r'[OIII]$\lambda$5007')
        error_oiii = popt_oiii[0]*popt_oiii[2]*np.sqrt((np.sqrt(pcov_oiii[0][0])/popt_oiii[0])**2+(np.sqrt(pcov_oiii[2][2])/popt_oiii[2])**2)
        oiii_flux = popt_oiii[0]*popt_oiii[2]
        #ax0.annotate('Flux = '+str(round(oiii_flux,1))+' +/- '+str(round(error_oiii,1)), xy=(0.02,0.87), xycoords='axes fraction')



        ax1 = fig.add_subplot(212)
        ax1.scatter(xs_hbeta, sum_hbeta, color='black')
        ax1.plot(xs_hbeta, gauss(xs_hbeta, *popt_hbeta), color='black')
        error_hbeta = popt_hbeta[0]*popt_hbeta[2]*np.sqrt((np.sqrt(pcov_hbeta[0][0])/popt_hbeta[0])**2+(np.sqrt(pcov_hbeta[2][2])/popt_hbeta[2])**2)
        hbeta_flux = popt_hbeta[0]*popt_hbeta[2]
        #ax1.annotate('Flux = '+str(round(hbeta_flux,1))+' +/- '+str(round(error_hbeta,1)), xy=(0.02,0.87), xycoords='axes fraction')

        ax1.set_title(r'H$\beta$')


        ratio_y = np.log10(np.array(popt_oiii[0]*popt_oiii[2])/np.array(popt_hbeta[0]*popt_hbeta[2]))
        #ratio_y_ll = np.log10(np.array(oiii_fit[4])/np.array(hbeta_fit[5]))
        dy = np.sqrt((np.array(error_oiii)/np.array(oiii_flux))**2+(np.array(error_hbeta)/np.array(hbeta_flux))**2)
        ratio_y_e = 0.434*(dy)


        ax0.annotate(r'log(OIII/H$\beta$) = '+str(round(ratio_y,2))+' +/- '+str(round(ratio_y_e,2)), xy=(0.55, 0.87), xycoords='axes fraction')

        if error_oiii > oiii_flux or error_hbeta > hbeta_flux:
            #then plot the lower limit
            ax1.axhline(y=popt_hbeta[3], color='red')
            ll = (popt_oiii[0]-popt_oiii[3])/popt_hbeta[3]
            ratio_ll = np.log10(np.array(ll))
            ax0.annotate('Lower limit = '+str(round(ratio_ll,2)), xy=(0.55, 0.77), xycoords='axes fraction')

        plt.tight_layout()
        ax0.set_ylim([np.min(sum_oiii)-0.2*np.max(sum_oiii),np.max(sum_oiii)+0.2*np.max(sum_oiii)])
        ax1.set_ylim([np.min(sum_hbeta)-0.2*np.max(sum_hbeta),np.max(sum_hbeta)+0.2*np.max(sum_hbeta)])

        plt.savefig('Adi_BPT/'+str(name)+'_PA_'+str(PA)+'_gaussian.png', dpi=500)
        return 0, 0, 0, 0
def gauss2(x,a,x0,sigma,offset,a2,x2,sigma2):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))+a2*np.exp(-(x-x2)**2/(2*sigma2**2))+offset
```
